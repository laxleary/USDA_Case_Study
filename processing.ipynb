## Web Scraping

```python
import requests
from bs4 import BeautifulSoup
import re
from urllib.parse import urljoin
import time
import json

# Base URL
main_url = "https://agcensus.library.cornell.edu/census_year/2012-census/"

# States list
states = [
    "United States", "Alabama", "Alaska", "Arizona", "Arkansas", "California",
    "Colorado", "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii",
    "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana",
    "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi",
    "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey",
    "New Mexico", "New York", "North Carolina", "North Dakota", "Ohio",
    "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina",
    "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia",
    "Washington", "West Virginia", "Wisconsin", "Wyoming"
]

# Headers to mimic a real browser
headers = {
    "User-Agent": "Chrome/137.0.0.0 Safari/537.36"
}

# Create a session to persist cookies and connections
session = requests.Session()
session.headers.update(headers)

# Step 1: Fetch main page
try:
    response = session.get(main_url)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
except Exception as e:
    print(f"Failed to fetch main page: {e}")
    exit(1)

# Step 2: Get all state page URLs
state_links = []

for td in soup.select("td.name.col-12"):
    link = td.find("a")
    if link and link.text.strip() in states:
        full_url = urljoin(main_url, link["href"])
        state_links.append({
            "state": link.text.strip(),
            "url": full_url
        })

print(f"Found {len(state_links)} state pages")

# Step 3: Visit each state page and get the PDF link
pdf_links = []

for entry in state_links:
    state = entry["state"]
    url = entry["url"]
    print(f"Fetching {state} page: {url}")

    try:
        res = session.get(url)
        res.raise_for_status()
        page_soup = BeautifulSoup(res.text, 'html.parser')

        download_tds = page_soup.select("td.download.col-2")
        pdf_url = None

        for td in download_tds:
            a_tag = td.find("a", attrs={
                "aria-label": re.compile(
                    r"Selected Operator Characteristics for Principal, Second, and Third Operator: 2012"
                )
            })
            if a_tag and a_tag.get("href"):
                pdf_url = urljoin(url, a_tag["href"])
                break  # first matching link only

        if pdf_url:
            print(f"{state}: {pdf_url}")
            pdf_links.append({"state": state, "pdf_url": pdf_url})
        else:
            print(f"{state}: Could not find matching PDF link")

    except Exception as e:
        print(f"{state}: Error fetching page - {e}")

    time.sleep(2)  # polite delay, adjust as needed


```

```python
import os

download_folder = 'raw_tables/2012/'
os.makedirs(download_folder, exist_ok=True)

for entry in pdf_links:
    state = entry["state"]
    pdf_url = entry["pdf_url"]

    # Create a safe filename (e.g. replace spaces with underscores)
    filename = f"{state.replace(' ', '_')}.pdf"
    filepath = os.path.join(download_folder, filename)

    # Skip download if file already exists
    if os.path.exists(filepath):
        print(f"Skipping {state}, already downloaded.")
        continue

    print(f"Downloading {state} PDF...")

    try:
        r = session.get(pdf_url, stream=True)
        r.raise_for_status()

        with open(filepath, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)
        print(f"Saved to {filepath}")
    except Exception as e:
        print(f"Failed to download {state}: {e}")

    time.sleep(2)  # polite delay between downloads to avoid connection closure
```

```python
import requests
from bs4 import BeautifulSoup
import re
from urllib.parse import urljoin
import time
import json

# Base URL
main_url = "https://agcensus.library.cornell.edu/census_year/2012-census/"

# States list
states = [
    "United States", "Alabama", "Alaska", "Arizona", "Arkansas", "California",
    "Colorado", "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii",
    "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana",
    "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi",
    "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey",
    "New Mexico", "New York", "North Carolina", "North Dakota", "Ohio",
    "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina",
    "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia",
    "Washington", "West Virginia", "Wisconsin", "Wyoming"
]

# Headers to mimic a real browser
headers = {
    "User-Agent": "Chrome/137.0.0.0 Safari/537.36"
}

# Create a session to persist cookies and connections
session = requests.Session()
session.headers.update(headers)

# Step 1: Fetch main page
try:
    response = session.get(main_url)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
except Exception as e:
    print(f"Failed to fetch main page: {e}")
    exit(1)

# Step 2: Get all state page URLs
state_links = []

for td in soup.select("td.name.col-12"):
    link = td.find("a")
    if link and link.text.strip() in states:
        full_url = urljoin(main_url, link["href"])
        state_links.append({
            "state": link.text.strip(),
            "url": full_url
        })

print(f"Found {len(state_links)} state pages")

# Step 3: Visit each state page and get the PDF link
pdf_links = []

for entry in state_links:
    state = entry["state"]
    url = entry["url"]
    print(f"Fetching {state}")

    try:
        res = session.get(url)
        res.raise_for_status()
        page_soup = BeautifulSoup(res.text, 'html.parser')

        download_tds = page_soup.select("td.download.col-2")
        pdf_url = None

        for td in download_tds:
            a_tag = td.find("a", attrs={
                "aria-label": re.compile(
                    r"Women Operators - Selected Operator Characteristics: 2012 and 2007"
                )
            })
            if a_tag and a_tag.get("href"):
                pdf_url = urljoin(url, a_tag["href"])
                break  # first matching link only

        if pdf_url:
            print(f"{state} data found.")
            pdf_links.append({"state": state, "pdf_url": pdf_url})
        else:
            print(f"{state}: Could not find matching PDF link")

    except Exception as e:
        print(f"{state}: Error fetching page - {e}")

    time.sleep(2)  # polite delay, adjust as needed

```

```python
import os

download_folder = 'raw_tables/2012/gender/'
os.makedirs(download_folder, exist_ok=True)

for entry in pdf_links:
    state = entry["state"]
    pdf_url = entry["pdf_url"]

    # Create a safe filename (e.g. replace spaces with underscores)
    filename = f"{state.replace(' ', '_')}_gendered_data.pdf"
    filepath = os.path.join(download_folder, filename)

    # Skip download if file already exists
    if os.path.exists(filepath):
        print(f"Skipping {state}, already downloaded.")
        continue

    print(f"Downloading {state} PDF...")

    try:
        r = session.get(pdf_url, stream=True)
        r.raise_for_status()

        with open(filepath, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)
        print(f"Saved to {filepath}")
    except Exception as e:
        print(f"Failed to download {state}: {e}")

    time.sleep(2)  # polite delay between downloads to avoid connection closure
```

```python
output_dict9 = {}
output_dict10 = {}

output_dict9['Producer Age'] = ['25 to 34 years', '35 to 44 years', '45 to 54 years', '55 to 64 years', '65 to 74 years', 'Average Age']
output_dict10['Producer Age'] = ['25 to 34 years', '35 to 44 years', '45 to 54 years', '55 to 64 years', '65 to 74 years', 'Average Age']

for state in states:
    output_dict9[state] = []
    output_dict10[state] = []

titles = {25: '25 to 34 years', 35:'35 to 44 years', 45:'45 to 54 years', 55:'55 to 64 years', 65:'65 to 74 years', 'AVG': 'Average Age'}
for age in titles.keys():
    doc = f'raw_tables/2022/female{age}.csv'
    df = pd.read_csv(doc)
    for state in states:
        if state == 'United States':
            title = 'US TOTAL'
        else:
            title = state.upper()
        #2022
        output_dict9[state].append(float(df.loc[df['State']==title].loc[:,'Value'].values[0].replace(',','')))
        #2017
        output_dict10[state].append(float(df.loc[df['State']==title].loc[:,'Value'].values[1].replace(',','')))
```
